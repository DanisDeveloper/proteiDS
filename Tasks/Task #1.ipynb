```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
filename = "../data/time_messagees.txt"
```

## Решение через библиотеку sklearn

```python
y = np.loadtxt(filename,delimiter=',',usecols=[1])
x = np.expand_dims(np.linspace(0,1,len(y)),1)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x)
```

```python
plt.figure(figsize=(12,8),dpi=200)
plt.plot(x,y_pred,c='r')
plt.scatter(x,y,s=1);
plt.xlabel('Время');
plt.ylabel('Количество сообщений');
labels = ['00:00','06:00','12:00','18:00','24:00']
plt.xticks(np.linspace(0,1,len(labels),dtype=float),labels=labels);
plt.legend(['Линейная регрессия','Количество сообщений']);
```

### Средняя квадратичная ошибка

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y,y_pred)
```

## Решение через статистические формулы

```python
y = np.loadtxt(filename,delimiter=',',usecols=[1])
x = np.linspace(0,1,len(y))
```

```python
slope = sum((x - x.mean())*(y - y.mean()))/sum((x - x.mean())**2)
slope
```

```python
bias = y.mean() - slope * x.mean()
bias
```

```python
y_pred = slope*x+bias
```

```python
plt.figure(figsize=(12,8),dpi=200)
plt.plot(x,y_pred,c='r')
plt.scatter(x,y,s=1);
plt.xlabel('Время');
plt.ylabel('Количество запросов');
labels = ['00:00','06:00','12:00','18:00','24:00']
plt.xticks(np.linspace(0,1,len(labels),dtype=float),labels=labels);
plt.legend(['Линейная регрессия','Количество сообщений']);
```

### Средняя квадратичная ошибка

```python
mean_squared_error(y,y_pred)
```
